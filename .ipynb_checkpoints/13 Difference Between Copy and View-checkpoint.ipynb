```python
 import numpy as np
```

## Copy

```python
var1 = np.array([1,2,3,4])
copy = var1.copy()
print(f'var1: {var1}')
print(f'copy: {copy}')

print(var1 is copy)
print(f'Address of var1 is {id(var1)}')
print(f'Address of copy is {id(copy)}')
print('Both are different')
```

```python
var1[2]=45
print(f'var1: {var1}')
print(f'copy: {copy}')
```

```python
copy[-1]=100
print(f'var1: {var1}') 
print(f'copy: {copy}')
```

## View

```python
var2 = np.array([9,8,7,6])
view = var2.view()
print(f'var2: {var2}')
print(f'view: {view}')
print(var2 is view)
print(f'Address of var2 {id(var2)}')
print(f'Address of view {id(view)}')
```

```python
var2[0] = 47
print(f'var2: {var2}')
print(f'view: {view}')
```

```python
view[-1]=106
print(f'var2: {var2}')
print(f'view: {view}')
```

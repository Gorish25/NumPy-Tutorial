 ## Join Array

```python
# 1D Array
var1 = np.array([1,2,3,4])
var2 = np.array([9,8,7,6])
var3 = np.concatenate((var1,var2))
print(var3)
```

```python
# 2D Array
var4 = np.array([[1,2],[3,4]])
var5 = np.array([[9,8],[7,6]])
print(var4)
print()
print(var5)
print()

# concatenating array depends on axis(0,1) (in terms of 2D)
# using axis=1
var6 = np.concatenate((var4,var5),axis=1)
print(var6)

print()

# using axis=0
var7 = np.concatenate((var4,var5),axis=0)
print(var7)
```

- using stack

```python
var8 = np.array([1,2,3,4])
var9 = np.array([9,8,7,6])

# along axis 0
print('using axis 0')
var10 = np.stack((var8,var9),axis=0)
print(var10)

print()

# along axis 1
print('using axis 1')
var11 = np.stack((var8,var9),axis=1)
print(var11)

print()

# using horizontal stack
print('using horizontal stack')
var12 = np.hstack((var8,var9)) # row
print(var12)

print()

# using vertical stack
print('using vertical stack')
var13 = np.vstack((var8,var9)) # column
print(var13)

print()

# along with height of stack
print('along with height')
var14 = np.dstack((var8,var9)) # height
print(var14)
```

## Split
- Splitting breaks one array into multiple

```python
# splitting 1D Array
var15 = np.array([1,2,3,4,5])
print(var15)

print()

# 1st arg is array 2nd arg is how many arrays we want
var16 = np.array_split(var15,3)
print(var16)
print(type(var16))
```

```python
# splitting 2D Array
var17 = np.array([[1,2],[3,4],[5,6]])
print(var17)

print()

var18 = np.array_split(var17,3)
print(var18)
print(type(var18))
```

```python
# splitting 2D arrays along with axis

# axis 0
print('Splitting along with axis 0')
var19 = np.array_split(var17,3,axis=0)
print(var19)

print()

# axis 1
print('Splitting along with axis 1')
var20 = np.array_split(var17,3,axis=1)
print(var20)
```
